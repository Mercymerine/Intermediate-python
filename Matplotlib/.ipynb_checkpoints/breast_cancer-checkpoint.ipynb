```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load breast cancer dataset
breast_cancer = load_breast_cancer()

#create features
features = breast_cancer.data

#create target
target = breast_cancer.target
```

```python
print(breast_cancer.DESCR)
```

```python
breast_cancer.feature_names
```

```python
df = pd.DataFrame(data = features, columns = breast_cancer.feature_names)
print(df.head())
```

```python
df
```

```python
df.dtypes
```

```python
plt.figure(figsize = (30, 10))
df_correlation = df.corr()

plt.title("A Heatmap Illustrating correlation among variables")
sns.heatmap(df_correlation, annot = True)
plt.show()
```

```python
features = ['worst perimeter', 'worst concave points', 'smoothness error']

plt.figure(figsize = (10, 10))
sns.pairplot(df[features])
plt.show()
```

```python
target_df = pd.Series(target, name = 'Target')
target_df
```

```python
plt.figure(figsize = (10, 10))
sns.countplot(target_df)
plt.title('Target Distribution')
plt.show()
```

```python

plt.figure(figsize = (89, 20))
df_sample = df[:5]
df_sample.plot(kind = 'bar')

plt.show()
```

```python
plt.pie(data = features)
plt.show()
```
