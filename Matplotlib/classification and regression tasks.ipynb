```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
#load mnist dataset.
mnist = load_digits()

#create features
features = mnist.data

#create target
target = mnist.target
```

```python
features = np.reshape(features, (features.shape[0], 8, 8))
features.shape
```

```python
plt.figure(figsize = (10, 10))
for item in range(16):
    plt.subplot(4,4, item + 1)
    plt.imshow(features[item])
    plt.title(target[item])
    
plt.suptitle('Samples of Mnist Dataset')
plt.axis('off')
plt.show()
```
